#Visulazing parts of speech

spacy offers an outstanding visualizer called displaCy:

```python
# Perform standard imports
import spacy
nlp = spacy.load('en_core_web_sm')

# Import the displaCy library
from spacy import displacy
```

```python
# Create a simple Doc object
doc = nlp(u"The quick  brown fox jumped over the lazy dog's back.")
```

```python
# render the dependency parse immediately inside Jupyter:
displacy.render(doc, style='dep', jupyter=True, options={'distance': 110})
```

## Creating Visualizations outside of jupyter

if you're using another Python IDE or writing a script, you can choose to have spacy serve up HTMP seperately.

instead od diplacy.render(), use displacy.serve():

```python
displacy.serve(doc, style ='dep', options={'distance': 110})   #http://127.0.0.1:5000/ ye alg se h 
```
